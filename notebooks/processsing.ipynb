```python
import os
import pandas as pd
import numpy as np
from PIL import Image
```

```python
PROJECT_DIR = os.path.dirname(os.getcwd())
DATA_DIR = os.path.join(PROJECT_DIR, 'data')
METADATA_PATH = os.path.join(DATA_DIR, 'metadata')
IMG_PATH = os.path.join(DATA_DIR, 'images')
```

```python
csv_path = os.path.join(METADATA_PATH, 'kaggle.xlsx')
metadata = pd.read_excel(csv_path)
metadata
```

```python
images_path = os.path.join(IMG_PATH, 'data_processed')
ids = []
for file_name in os.listdir(images_path):
    # print(os.path.join(images_path, file_name))
    id = os.path.splitext(file_name)[0]
    ids.append(id)
```

```python
img = os.listdir(images_path)[10]
image = os.path.join(images_path, img)
```

```python
clean_data_dict = {}
for img in os.listdir(images_path)[:200]:
    if ".jpg" in img:
        image = os.path.join(images_path, img)
        clean_data_dict[img.replace('.jpg', '')] = np.array(Image.open(image))
    else:
        continue
clean_data_df = pd.DataFrame(clean_data_dict.items(), columns=['image_id', 'np_array_img'])
```

```python
final_sample = clean_data_df.combine_first(metadata[['image_id', 'cancer']])
final_sample_train = pd.DataFrame(final_sample.head(200))
```
